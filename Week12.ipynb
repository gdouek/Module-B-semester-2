# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
from IPython.display import Markdown, display

display(Markdown("""
**Weekly graph question**

The first option, with the two weeks shown in **separate side-by-side plots**, has the advantage that each week is visually isolated. This can make it easier to focus on the shape of each individual week without confusion. It is also helpful if the scales or patterns are very different and we want to compare the overall trends separately.

However, the disadvantage is that it is harder to compare the two weeks directly, because the reader has to look back and forth between the two panels.

The second option, with both weeks shown on the **same plot**, has the advantage that it makes direct comparison much easier. We can quickly see where one week had more visitors than the other and compare the trends day by day.

The disadvantage is that the graph can become more cluttered, especially if the lines are similar or if there are many time series. In that case, it may be harder to read.

Overall, the separate plots are better for emphasizing the shape of each series individually, while the combined plot is better for direct comparison between the two weeks.
"""))
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
week1 = [120, 135, 150, 160, 180, 200, 210]
week2 = [110, 130, 140, 170, 220, 260, 300]

plt.figure(figsize=(10,5))

# Plot both lines
plt.plot(days, week1, linewidth=2)
plt.plot(days, week2, linewidth=3)

# Highlight last point of week2
plt.scatter(days[-1], week2[-1], s=80)

# Remove clutter
sns.despine()

# Titles
plt.title("Weekly Visitors Comparison", loc="left", fontsize=14, weight="bold")
plt.ylabel("Visitors")

# Annotation (key storytelling element)
plt.annotate(
    "Big increase in Week 2",
    xy=(6, week2[-1]),
    xytext=(4.5, week2[-1] - 80),
    arrowprops=dict(arrowstyle="->")
)

plt.show()
```
