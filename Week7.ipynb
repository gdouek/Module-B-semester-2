```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

## Weekly Graph Question

If the goal is to direct the viewer’s attention to the **last three months of data**, we can use **preattentive attributes** to guide the viewer’s eye to that portion of the graph first. Preattentive attributes are visual properties that the brain processes very quickly, such as color, size, position, and contrast.

Several techniques could be used:

### 1. Use Color to Highlight the Last Three Months
The last three months could be displayed in a **different color** (for example, red or orange) while the earlier months are shown in a **lighter gray**. This creates contrast and immediately draws attention to the highlighted data.

### 2. Increase Line Thickness or Marker Size
The data points for the last three months could be plotted with **larger markers or thicker lines**, making them visually stand out compared to the rest of the time series.

### 3. Add Annotations or Labels
Text annotations could be placed near the final data points to label the values or highlight the recent changes. This helps guide the viewer’s attention to the most important portion of the chart.

### 4. Use Shading or Background Highlighting
A light shaded region could be added behind the final three months of the timeline. This visually separates that section from the rest of the chart and naturally draws attention to it.

### 5. Reduce Emphasis on Earlier Data
The earlier months could be displayed using **lighter colors or lower opacity**, while the last three months remain darker and more prominent. This technique helps the viewer focus on the most recent data first.

Using these preattentive attributes helps control the visual hierarchy of the chart and directs the viewer’s attention to the most important part of the data.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate random data
np.random.seed(42)

categories = [
    "Category A",
    "Category B",
    "Category C",
    "Category D",
    "Category E",
    "Category F"
]

values = np.random.randint(20, 100, size=len(categories))

data = pd.Series(values, index=categories).sort_values()

# Colors: gray for all, highlight the largest
colors = ["lightgray"] * len(data)
colors[-1] = "steelblue"

plt.figure(figsize=(10,5))
data.plot(kind="barh", color=colors)

plt.title("Example Storytelling With Data Chart", fontsize=14)
plt.xlabel("Value")

# Remove chart junk
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.grid(axis="x", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```
